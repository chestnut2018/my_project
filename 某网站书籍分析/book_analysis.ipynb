```python
data[data['页数'].isnull()]
```

```python
def can_fill(col_name):
    try:
        nan_rows = data[col_name].isnull().sum()
        a = nan_rows/len(data)
        if a< 0.05:
            return '共有{}个空值，空值占比{:.4f},可以进行补值操作'.format(nan_rows,a)
        else:
            return '空值太多了，没有补的价值，共有{}个空值,空值占比{:.4f}'.format(nan_rows,a)
    except:
        return '这列没有空值' 
```

```python
def can_fill(col_name):
    try:
        nan_rows = data2[col_name].isnull().sum()
        a = nan_rows/len(data2)
        if a< 0.05:
            return '共有{}个空值，空值占比{:.4f},可以进行补值操作'.format(nan_rows,a)
        else:
            return '空值太多了，没有补的价值，共有{}个空值,空值占比{:.4f}'.format(nan_rows,a)
    except:
        return '这列没有空值' 
```

```python
can_fill('价格')
```

```python
def mean_std_cv(col_name):
    mean = data2[col_name].mean(skipna=True)
    std = data2[col_name].std(skipna=True)
    cv = std/mean
    return '平均值{},标准差{},变异系数{}'.format(mean,std,cv)
```

```python
mean_std_cv('价格')
```

变异系数过大，且稀疏，用bfill或fill补值

```python
def nan_fill_bf(col_name):
    col = data2[col_name].copy()
    col = col.fillna(method='bfill')
    if col.isnull().value_counts()[False] != col.shape[0]:
        col = col.fillna(method='ffill')
    return col
```

```python
data2['价格'] = nan_fill_bf('价格')
```

### 评分高低与评分数量之间的关系

```python
data4['评分'].describe()
```

```python
group =[0,5,6,7,8,9,15]
```

```python
score.value_counts()
```

```python
score.codes
```

```python
data4['评分区间'] = score.codes.tolist()
```

由图可见，在5分到9分之间，符合评分越高，评论数量越多
